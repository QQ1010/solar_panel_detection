```python
import folium
from folium.plugins import MarkerCluster
import json
import numpy as np
```

```python
myMap = folium.Map([23.450648316816473,120.40191650390626], zoom_start=10)
myMap
```

```python
folium.GeoJson(
    'result.json'
).add_to(myMap)
myMap
```

```python
marker_cluster = MarkerCluster()
folium.LayerControl().add_to(myMap);
with open('result.json','r') as f:
    data = json.load(f)
for feature in data['features']:
    poly = feature['geometry']['coordinates'][0]
    x,y = np.mean(np.array(poly),axis=0)
#     myMap.add_child(child=folium.Marker(location=[y,x]))
    folium.Marker(location=[y,x]).add_to(marker_cluster)

myMap
```

```python
# 自動轉檔成html
myMap.save('myMap3.html')
```
